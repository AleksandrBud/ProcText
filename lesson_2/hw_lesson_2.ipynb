    â€¢ Ð˜ÑÐºÐ»ÑŽÑ‡Ð¸Ð¼ ÑÑ‚Ð¾Ð¿-ÑÐ»Ð¾Ð²Ð° Ñ Ð¿Ð¾Ð¼Ð¾Ñ‰ÑŒÑŽ stop_words='english'.
    â€¢ ÐžÑ‚Ð¾Ð±Ñ€Ð°Ð·Ð¸Ð¼ Bag-of-Words Ð¼Ð¾Ð´ÐµÐ»ÑŒ ÐºÐ°Ðº DataFrame. columns Ð½ÐµÐ¾Ð±Ñ…Ð¾Ð´Ð¸Ð¼Ð¾ Ð¸Ð·Ð²Ð»ÐµÑ‡ÑŒ Ñ Ð¿Ð¾Ð¼Ð¾Ñ‰ÑŒÑŽ CountVectorizer.get_feature_names().


```python
def vectorize(tokens):
    model = CountVectorizer(stop_words='english',
                                   max_features=1000, 
                                   max_df=0.9,)
    documents =[]
    for token in tokens:
        for word in token:
            documents.append(word)
  
    fit_model = model.fit_transform(documents)
    
    return model, fit_model, model.get_feature_names()
```

```python
count_vectorizer_stem, bag_of_words_stem, feature_names_stem = vectorize(combine_df['tweet_stemmed'].values)
count_vectorizer_lemm, bag_of_words_lemm, feature_names_lemm = vectorize(combine_df['tweet_lemmatized'].values)
```

```python
pd.DataFrame(bag_of_words_stem.toarray(), columns = feature_names_stem)
```

```python
pd.DataFrame(bag_of_words_lemm.toarray(), columns = feature_names_lemm)
```

```python
def vectorize_tfid(tokens):
    model = TfidfVectorizer(stop_words='english',
                                   max_features=1000, 
                                   max_df=0.9,)
    documents =[]
    for token in tokens:
        for word in token:
            documents.append(word)  
    fit_model = model.fit_transform(documents)
    
    return model, fit_model, model.get_feature_names()
```

```python
count_vectorizer_stem_t, bag_of_words_stem_t, feature_names_stem_t = vectorize_tfid(combine_df['tweet_stemmed'].values)
count_vectorizer_lemm_t, bag_of_words_lemm_t, feature_names_lemm_t = vectorize_tfid(combine_df['tweet_lemmatized'].values)
```

```python
from gensim.models import Word2Vec
```

```python
modelW2V = Word2Vec(size=200,
                    window=5,
                    min_count=2,
                    sg = 1,
                    hs = 0,
                    negative = 10,
                    workers= 32,
                    seed = 34)
```

```python
modelW2V.build_vocab(combine_df['tweet_token'].values)
```

```python
modelW2V.train(combine_df['tweet_token'].values, 
                total_examples=combine_df['tweet_token'].shape[0], 
                epochs=20,
               )
```
